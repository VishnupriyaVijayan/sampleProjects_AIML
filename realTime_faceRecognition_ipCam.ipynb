```python
import cv2
import os
import numpy as np
import urllib.request
import imutils
import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning)
```

```python

haar_file = 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(haar_file)
datasets = 'datasets'

# Initialize variables
images = []
labels = []
names = {}
id = 0
(width, height) = (130, 100)
```

```python
for (subdirs, dirs, files) in os.walk(datasets):
    for subdir in dirs:
        names[id] = subdir
        subjectpath = os.path.join(datasets, subdir)
        for filename in os.listdir(subjectpath):
            path = os.path.join(subjectpath, filename)
            label = id
            images.append(cv2.imread(path, 0))
            labels.append(int(label))
        id += 1

# Specify dtype=object when creating the NumPy array
(images, labels) = [np.array(lis) for lis in [images, labels]]

```

```python

model =cv2.face_LBPHFaceRecognizer.create()
#model =  cv2.face.FisherFaceRecognizer_create()
model.train(images, labels)
```

```python



ip_camera_url =  'http://192.168.1.35:8080/shot.jpg'

while True:
    # Capture image from the Android camera
    imgPath = urllib.request.urlopen(ip_camera_url)
    imgNp = np.array(bytearray(imgPath.read()), dtype=np.uint8)
    img = cv2.imdecode(imgNp, -1)

    # Flip the image vertically
    img = cv2.flip(img, 0)

    # Resize the image for better processing
    im = imutils.resize(img, width=400)

    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(im, (x, y), (x + w, y + h), (255, 255, 0), 2)
        face = gray[y:y + h, x:x + w]
        face_resize = cv2.resize(face, (width, height))

        prediction = model.predict(face_resize)
        cv2.rectangle(im, (x, y), (x + w, y + h), (0, 255, 0), 3)

        if prediction[1] < 800:
            cv2.putText(im, '%s - %.0f' % (names[prediction[0]], prediction[1]), (x-10, y-10),
                        cv2.FONT_HERSHEY_PLAIN, 1, (255, 0, 100),2)
            print(names[prediction[0]])
        else:
            cv2.putText(im, 'Unknown', (x-10, y-10), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 0))

    cv2.imshow('FaceRecognition', im)
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break

cv2.destroyAllWindows()
```
