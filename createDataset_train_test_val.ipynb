```python


# Set the paths
source_folder = 'simple_images'
destination_folder = 'Datasets'

# Create train, validation, and test folders
train_folder = os.path.join(destination_folder, 'train')
val_folder = os.path.join(destination_folder, 'val')
test_folder = os.path.join(destination_folder, 'test')

os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# List of subfolders (classes)
classes = ['cat', 'fox']

# Set the ratios for train, validation, and test
train_ratio = 0.75
val_ratio = 0.15
test_ratio = 0.15

# Iterate over each class
for class_name in classes:
    class_source_folder = os.path.join(source_folder, class_name)
    class_train_folder = os.path.join(train_folder, class_name)
    class_val_folder = os.path.join(val_folder, class_name)
    class_test_folder = os.path.join(test_folder, class_name)

    os.makedirs(class_train_folder, exist_ok=True)
    os.makedirs(class_val_folder, exist_ok=True)
    os.makedirs(class_test_folder, exist_ok=True)

    # List all files in the class folder
    files = os.listdir(class_source_folder)
    random.shuffle(files)

    # Calculate the split indices
    train_split = int(train_ratio * len(files))
    val_split = int((train_ratio + val_ratio) * len(files))

    # Copy files to the corresponding folders
    for i, file in enumerate(files):
        source_path = os.path.join(class_source_folder, file)
        if i < train_split:
            destination_path = os.path.join(class_train_folder, file)
        elif i < val_split:
            destination_path = os.path.join(class_val_folder, file)
        else:
            destination_path = os.path.join(class_test_folder, file)
        
        shutil.copyfile(source_path, destination_path)

print("Dataset split into train, val, and test sets.")
```
